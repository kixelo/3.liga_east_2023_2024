```python
from bs4 import BeautifulSoup
import requests
import csv
import pandas as pd

url = "https://sportnet.sme.sk/futbalnet/z/sfz/s/4408/tabulky/?part=647a3f3b76d0d348cd095fa9&tableView="

headers = {'User_Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/124.0.0.0 Safari/537.36'} # in web search type: "what is my user agent"

response = requests.get(url, headers=headers)
response
```

```python
soup = BeautifulSoup(response.content, 'html.parser')
soup
```

```python
datas = soup.find_all('tr', class_='sc-Qpmqz sc-qiJqY cQeJtZ bJMciK' )
datas
```

```python
R = []
T = []
W = []
D = []
L = []
GS = []
GA = []
for data in datas: 
    rank = data.find('td', class_='sc-jVKKMF sc-hsnHQP iaisEp iaAxLF').text
    R.append(rank)
    
    team = data.find('div', class_='sc-ieHAqa cnUItO').text
    T.append(team)
    
    win = data.find_all('td', class_='sc-jVKKMF iaisEp')[1].text
    W.append(win)
    
    draw = data.find_all('td', class_='sc-jVKKMF iaisEp')[2].text
    D.append(draw)
    
    loss = data.find_all('td', class_='sc-jVKKMF iaisEp')[3].text
    L.append(loss)
    
    gs  = data.find_all('td', class_='sc-jVKKMF iaisEp')[4].text.split(":")[0]
    GS.append(gs)
    
    ga =  data.find_all('td', class_='sc-jVKKMF iaisEp')[4].text.split(":")[1]
    GA.append(ga)
    
df = pd.DataFrame({"Ranking": R, "Teams": T, "Wins": W, "Draws": D, "Losses": L, "Goals scored": GS, "Goals against": GA})
df.to_csv("3_liga_east_2023_24.csv", encoding='utf-8-sig', index=False)
```
